```python
# comparing sizes of both datasets to figure out what percentage of data is being lost
# when the subset is being taken
# need to specify the timestamp variable to get the full size of the dataset (not just the number of rows w/ the last column value not a NaN)
size_df_sample = df_sample.timestamp.size
print(f"Size of the sample dataset: {size_df_sample}")
size_df_Rrocket = df_Rrocket.timestamp.size
print(f"Size of the Rrocket dataset: {size_df_Rrocket}")
percentage_kept = (size_df_sample/size_df_Rrocket) * 100
print(f"The percentage of the data being kept in the subset is {percentage_kept}%")
percentage_lost = 100 - percentage_kept
print(f"The percentage of the data lost bc of the subset is {percentage_lost}%")
```

Just keeping the first few data values equal to the size of the sample dataset

```python
# keeps the first few data values equal to the size of the sample dataset
df_Rrocket_subset_start = df_Rrocket[0:size_df_sample]
df_Rrocket_subset_start
```

```python
# saves that data subset dataframe into a new .csv file
df_Rrocket_subset_start.to_csv('/content/drive/MyDrive/Subset dataset code/Rrocket_subset_start/events.csv', index = False)
```

```python
# keeps the last few data values equal to the size of the sample dataset
first_end_index = size_df_Rrocket - size_df_sample
df_Rrocket_subset_end = df_Rrocket[first_end_index:]
df_Rrocket_subset_end
```

```python
# saves that data subset dataframe into a new .csv file
df_Rrocket_subset_end.to_csv('/content/drive/MyDrive/Subset dataset code/Rrocket_subset_end/events.csv', index = False)
```

```python
# randomly sample through the dataset and retaining index order
df_Rrocket_subset_random = df_Rrocket.sample(n=size_df_sample, random_state=1).sort_index()
df_Rrocket_subset_random = df_Rrocket_subset_random.reset_index(drop = True)
df_Rrocket_subset_random
```

```python
# saves that data subset dataframe into a new .csv file
df_Rrocket_subset_random.to_csv('/content/drive/MyDrive/Subset dataset code/Rrocket_subset_random/events.csv', index = False)
```
